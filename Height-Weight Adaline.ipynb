```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from numpy.linalg import inv
```

```python
N = 100
std = 10
def BMI(height,weight):
    BMI = weight / (height/100)
    if BMI <= 18.4:
        return -1
    elif BMI <= 24.9:
        return 1
    elif BMI <= 29.9:
        return 1
    elif BMI <= 34.9:
        return 1
    elif BMI <= 39.9:
        return -1
    else:
        return -1
def generate_data(N):
    X = []
    Y = []
    n=0
    while n != N:
    #for i in range(N):
        x = np.random.randint(1, 175)
        y = 4 * np.random.randint(1, 175)
        
        if BMI(x,y) == 1:
            X.append(x)
            Y.append(y)
            n+=1

    X = np.array(X)
    Y = np.array(Y)

    X = X.reshape(N,1)
    Y = Y.reshape(N,1)
    return X,Y
    
X_train,Y_train = generate_data(100)
print(np.shape(X_train))
plt.scatter(X_train,Y_train)
```

```python
class AdalineRegressor:
    def __init__(self):
        pass
    
    
    def fit(self, X_train, Y_train):
        self.w = np.matmul(inv(np.matmul(X_train.T,X_train)), np.matmul(X_train.T, Y_train))
    
    
    def predict(self, X_test):
        Y_pred = np.matmul(X_test, self.w)
        
        return Y_pred
    
    
    def evaluate(self, X_test, Y_test):
        Y_pred = np.matmul(X_test, self.w)
        sub = np.abs(X_test - Y_test)
        avg = np.mean(sub)
        
        print(avg)
```
